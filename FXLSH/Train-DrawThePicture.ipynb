```python
N = 50
x = np.linspace(0, 10, N)
from numpy.random import rand
e = rand(N)*5.0 # noise
y1 = x + e
areas = rand(N)*300
plt.scatter(x, y1, s=areas)
colors = rand(N)
plt.scatter(x, y1, s=areas,c=colors)
plt.colorbar()
plt.title("Random scatter")
```

```python
# Create some data
e1 = rand(100)
e2 = rand(100)*2
e3 = rand(100)*10
e4 = rand(100)*100
corrmatrix = \
np.corrcoef([e1, e2,e3, e4])
# Plot corr matrix as image
plt.imshow(corrmatrix,cmap='GnBu')
plt.colorbar()
```

```python
t = np.linspace(0, 2*np.pi)
x = np.sin(t)
y = np.cos(t)
# To divide the plotting area
plt.subplot(2, 1, 1)
plt.plot(x)
# Now activate a new plot
# area.
plt.subplot(2, 1, 2)
plt.plot(y)
```

```python
# Create array of data
from numpy.random import randint
data = randint(10000,size=(10,1000))
# Approx norm distribution
x = np.sum(data, axis=0)
# Set up for stacked plots
plt.subplot(2,1,1)
plt.hist(x, color='r')
# Plot cumulative dist
plt.subplot(2,1,2)
plt.hist(x, cumulative=True)
# For multiple histograms:
```
